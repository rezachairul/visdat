```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = "Dataset2-consumer-behavior.csv"
df = pd.read_csv(file_path)
```

```python
# Variable Transformation
# 1. Menghapus simbol '$' dan mengubah ke numerik
df['Purchase_Amount'] = df['Purchase_Amount'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
# 2. Mengonversi Time_of_Purchase ke format datetime
df['Time_of_Purchase'] = pd.to_datetime(df['Time_of_Purchase'])
```

```python
# 3. Mengubah Discount_Used dan Customer_Loyalty_Program_Member ke format numerik (0 dan 1)
df['Discount_Used'] = df['Discount_Used'].map({'Yes': 1, 'No': 0})
df['Customer_Loyalty_Program_Member'] = df['Customer_Loyalty_Program_Member'].map({'Yes': 1, 'No': 0})
```

```python
# Menampilkan hasil transformasi
print(df.head())
```

```python
# Simpan hasil ke file baru jika diperlukan
df.to_csv("processed_dataset.csv", index=False)
```

```python
# Visualisasi 1: Bar Chart - Jumlah pelanggan berdasarkan kategori usia
plt.figure(figsize=(10, 5))
sns.countplot(x='Age', data=df, palette='viridis')
plt.title('Jumlah Pelanggan Berdasarkan Kategori Usia')
plt.xlabel('Usia')
plt.ylabel('Jumlah Pelanggan')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualisasi 2: Histogram - Distribusi Jumlah Pembelian
plt.figure(figsize=(10, 5))
sns.histplot(df['Purchase_Amount'], bins=20, kde=True, color='blue')
plt.title('Distribusi Jumlah Pembelian (Purchase Amount)')
plt.xlabel('Jumlah Pembelian ($)')
plt.ylabel('Frekuensi')
plt.show()
```
